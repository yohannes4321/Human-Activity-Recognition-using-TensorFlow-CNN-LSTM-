# Process the Dataset

```python
IMAGE_HEIGHT , IMAGE_WIDTH = 64, 64
SEQUENCE_LENGTH =20
DATASET_DIR = "UCF50"
CLASSES_LIST = ["WalkingWithDog", "TaiChi", "Swing", "HorseRace"]
```

```python
def frames_extraction(video_path):
  frames_list=[]
  video_reader=cv2.VideoCapture(video_path)
  video_frames_count=int(video_reader.get(cv2.CAP_PROP_FRAME_COUNT))
  skip_frames_window=max(int(video_frames_count/SEQUENCE_LENGTH),1)
  for frame_counter in range(SEQUENCE_LENGTH):
    video_reader.set(cv2.CAP_PROP_POS_FRAMES,frame_counter*skip_frames_window)
    success,frame=video_reader.read()
    if not success:
      break
    resized_frame=cv2.resize(frame,(IMAGE_HEIGHT,IMAGE_WIDTH))
    normalized_frame=resized_frame/255
    frames_list.append(normalized_frame)
  video_reader.release()
  return frames_list
```

```python
def create_dataset():
  fetures=[]
  lables=[]
  video_files_paths=[]
  for class_index,class_name in enumerate(CLASSES_LIST):
    print(f"Extracting Data of Class: {class_name}")
    files_list=os.listdir(os.path.join(DATASET_DIR,class_name))
    for file_name in files_list:
      video_file_path=os.path.join(DATASET_DIR,class_name,file_name)
      frames=frames_extraction(video_file_path)
      if len(frames)==SEQUENCE_LENGTH:
        fetures.append(frames)
        lables.append(class_index)
        video_files_paths.append(video_file_path)
  fetures=np.asarray(fetures)
  lables=np.array(lables)
  return fetures,lables,video_files_paths
```

```python
features,lables,video_files_paths=create_dataset()
```

```python
one_hot_encoded_lables=to_categorical(lables)
```

## Split the Data into Trian and Test Set

```python
features_train,features_test,lables_train,lables_test=train_test_split(features,one_hot_encoded_lables,test_size=0.25,shuffle=True,random_state=seed_constant)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import ConvLSTM2D, MaxPooling3D, TimeDistributed, Dropout, Flatten, Dense

def create_conv_lstm_model():
    model = Sequential()

    # 1st ConvLSTM2D Layer
    model.add(ConvLSTM2D(
        filters=4,
        kernel_size=(3, 3),
        activation='tanh',
        recurrent_dropout=0.2,
        return_sequences=True,
        input_shape=(SEQUENCE_LENGTH, IMAGE_HEIGHT, IMAGE_WIDTH, 3)
    ))
    model.add(MaxPooling3D(pool_size=(1, 2, 2), padding='same'))
    model.add(TimeDistributed(Dropout(0.2)))

    # 2nd ConvLSTM2D Layer
    model.add(ConvLSTM2D(filters=8, kernel_size=(3, 3), activation='tanh', recurrent_dropout=0.2, return_sequences=True))
    model.add(MaxPooling3D(pool_size=(1, 2, 2), padding='same'))
    model.add(TimeDistributed(Dropout(0.2)))

    # 3rd ConvLSTM2D Layer
    model.add(ConvLSTM2D(filters=14, kernel_size=(3, 3), activation='tanh', recurrent_dropout=0.2, return_sequences=True))
    model.add(MaxPooling3D(pool_size=(1, 2, 2), padding='same'))
    model.add(TimeDistributed(Dropout(0.2)))

    # 4th ConvLSTM2D Layer
    model.add(ConvLSTM2D(filters=16, kernel_size=(3, 3), activation='tanh', recurrent_dropout=0.2, return_sequences=True))
    model.add(MaxPooling3D(pool_size=(1, 2, 2), padding='same'))
    model.add(TimeDistributed(Dropout(0.2)))

    # Flatten and Fully Connected Layers
    model.add(Flatten())
    model.add(Dense(len(CLASSES_LIST), activation='softmax'))

    # Print the summary of the model
    model.summary()

    return model
```

```python
convlstm_model=create_conv_lstm_model()
print("Model successfully Created ")
```

# Cheack the Model **Structure**

```python
EarlyStopping_callback=EarlyStopping(monitor='val_loss',patience=15,mode='min',restore_best_weights=True)
convlstm_model.compile(loss='categorical_crossentropy',optimizer='Adam',metrics=['accuracy'])
convlstm_model_training_history=convlstm_model.fit(x=features_train,y=lables_train,epochs=5,batch_size=4,shuffle=True,validation_split=0.2,callbacks=[EarlyStopping_callback])
#
```

```python
model_evaluation_history=convlstm_model.evaluate(features_test,lables_test)
```

```python
model_evaluation_loss,model_evaluation_accuracy=model_evaluation_history
date_time_format='%Y_%m_%d__%H_%M_%S'
current_date_time_dt=dt.datetime.now()
current_date_time_string=dt.datetime.strftime(current_date_time_dt,date_time_format)
model_file_name=f'convlstm_model___Date_Time_{current_date_time_string}___loss_{model_evaluation_loss}___accuracy_{model_evaluation_accuracy}.h5'
convlstm_model.save(model_file_name)
```

```python
def plot_metric(model_training_history,metric_name_1,metric_name_2,plot_name):
  metric_value_1=model_training_history.history[metric_name_1]
  metric_value_2=model_training_history.history[metric_name_2]
  epochs=range(len(metric_value_1))
  plt.plot(epochs,metric_value_1,'blue',label=metric_name_1)
  plt.plot(epochs,metric_value_2,'red',label=metric_name_2)
  plt.title(str(plot_name))
  plt.legend()
```

```python
plot_metric(convlstm_model_training_history,'loss','val_loss','Total Loss vs Total Validation Loss')
```

```python
# Define constants for input shape
SEQUENCE_LENGTH = 20  # Example: Number of frames in a video sequence
IMAGE_HEIGHT = 64     # Example: Height of each frame
IMAGE_WIDTH = 64      # Example: Width of each frame

# Example class list: Modify this to match your actual classes
CLASSES_LIST = ["Class1", "Class2", "Class3", "Class4"]  # Replace with your class names

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import TimeDistributed, Conv2D, MaxPooling2D, Dropout, Flatten, LSTM, Dense

def create_LRCN_model():
    model = Sequential()

    # TimeDistributed Conv2D Layers for Feature Extraction
    model.add(TimeDistributed(Conv2D(16, (3, 3), padding='same', activation='relu'), input_shape=(SEQUENCE_LENGTH, IMAGE_HEIGHT, IMAGE_WIDTH, 3)))
    model.add(TimeDistributed(MaxPooling2D((4, 4))))
    model.add(TimeDistributed(Dropout(0.25)))

    model.add(TimeDistributed(Conv2D(32, (3, 3), padding='same', activation='relu')))
    model.add(TimeDistributed(MaxPooling2D((4, 4))))
    model.add(TimeDistributed(Dropout(0.25)))

    model.add(TimeDistributed(Conv2D(64, (3, 3), padding='same', activation='relu')))
    model.add(TimeDistributed(MaxPooling2D((2, 2))))
    model.add(TimeDistributed(Dropout(0.25)))

    model.add(TimeDistributed(Conv2D(64, (3, 3), padding='same', activation='relu')))
    model.add(TimeDistributed(MaxPooling2D((2, 2))))
    model.add(TimeDistributed(Dropout(0.25)))

    # Flatten the output before passing to LSTM
    model.add(TimeDistributed(Flatten()))

    # LSTM Layer for Sequential Learning
    model.add(LSTM(32))
    model.add(Dropout(0.25))

    # Fully Connected Layer for Classification
    model.add(Dense(len(CLASSES_LIST), activation='softmax'))

    # Print Model Summary
    model.summary()

    return model

# Create the model
LRCN_model = create_LRCN_model()
print("Model successfully Created")
```

```python
from pytube import YouTube
import os

def download_video(video_url, output_path='videos'):
    # Create the output directory if it doesn't exist
    if not os.path.exists(output_path):
        os.makedirs(output_path)

    # Download the video
    try:
        yt = YouTube(video_url)
        video = yt.streams.filter(progressive=True, file_extension='mp4').first()  # Choose the first available stream
        video.download(output_path)
        print(f"Downloaded: {video.title}")
    except Exception as e:
        print(f"Error downloading video: {e}")

# Example video URL (replace this with your desired video URL)
video_url = 'https://www.youtube.com/watch?v=VIDEO_ID'  # Replace VIDEO_ID with the actual ID
download_video(video_url)
```

```python
import cv2

def check_video_properties(video_path):
    # Open the video file
    cap = cv2.VideoCapture(video_path)

    if not cap.isOpened():
        print("Error opening video file")
        return

    # Get video properties
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = cap.get(cv2.CAP_PROP_FPS)

    print(f"Frame Count: {frame_count}")
    print(f"Width: {width}")
    print(f"Height: {height}")
    print(f"FPS: {fps}")

    # Release the video capture object
    cap.release()

# Path to the downloaded video (adjust the filename as needed)
video_file_path = os.path.join('videos', 'DownloadedVideo.mp4')  # Replace with the actual downloaded filename
check_video_properties(video_file_path)
```
